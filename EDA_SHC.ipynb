```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def station_arrival(file_dir, file_name, station, output):
    
    for i in range(len(file_name)):
        
        string = file_dir + file_name[i] + ".xlsx"
        temp = pd.read_excel(string, skiprows = 1)
        temp = temp.set_index('Unnamed: 0')
        output.loc[file_name[i][-6:]] = temp.loc['Entries'][station]
    
    return output
```

```python
# Ridership 2022

file_dir = "Data/ridership_2022/"
file_name = ['Ridership_202201', 'Ridership_202202', 'Ridership_202203', 'Ridership_202204', 'Ridership_202205', 'Ridership_202206', 
             'Ridership_202207', 'Ridership_202208', 'Ridership_202209', 'Ridership_202210', 'Ridership_202211', 'Ridership_202212']
station = ['EM','MT', 'PL', 'CC']
Ridership_2022 = pd.DataFrame(columns = station)

Ridership_2022 = station_arrival(file_dir, file_name, station, Ridership_2022)
Ridership_2022
```

```python
Ridership_2022.plot()

plt.legend(['EM', 'MT', 'PL', 'CC'])
plt.title('2022 SF CBD BART Station Weekday Arrival Counts')
plt.xlabel('Month')
plt.ylabel('Average arrival counts')
```
